```python
#filling missing values in the categorical columns with mode
data.Gender=data.Gender.fillna(data.Gender.mode()[0])
data.Married=data.Married.fillna(data.Married.mode()[0])
data.isna().sum()
```

```python
data.Self_Employed.unique()
```

```python
data.Dependents=data.Dependents.fillna(data.Dependents.mode()[0])
data.Self_Employed=data.Self_Employed.fillna(data.Self_Employed.mode()[0])
data.isna().sum()
```

```python
data.LoanAmount=data.LoanAmount.fillna(data.LoanAmount.median())
```

```python
data.Loan_Amount_Term.unique()
```

```python
data.Loan_Amount_Term=data.Loan_Amount_Term.fillna(data.Loan_Amount_Term.median())
```

```python
data.Credit_History.unique()
```

```python
#filling of missing values in the numerical columns using median
data.LoanAmount=data.LoanAmount.fillna(data.LoanAmount.median())
data.Loan_Amount_Term=data.Loan_Amount_Term.fillna(data.Loan_Amount_Term.median())
data.Credit_History=data.Credit_History.fillna(data.Credit_History.median())
data.isna().sum()
```

C.Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.Gender=le.fit_transform(data.Gender)
data.Married=le.fit_transform(data.Married)
data.Education=le.fit_transform(data.Education)
data.Self_Employed=le.fit_transform(data.Self_Employed)
data.Property_Area=(le.fit_transform(data.Property_Area))
data.Loan_ID=le.fit_transform(data.Loan_ID)
data.Credit_History=le.fit_transform(data.Credit_History)
data.Dependents=le.fit_transform(data.Dependents)
data.head()
```

```python
data.dtypes
```

d. Scaling

```python
x=data.drop('Loan_Status',axis=1)
y=data.Loan_Status
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
num_cols=['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
x.loc[:,num_cols]=ss.fit_transform(x[num_cols])
x.head()
```

Ensemble Learning Techniques

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_rf)
```

```python
#Bagging Classifier
from  sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(n_estimators=100,random_state=42)
bg.fit(x_train,y_train)
y_pred_bg=bg.predict(x_test)
accuracy_score(y_test,y_pred_bg)
```

```python
#Bagging classifier using SVM
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
bgg=BaggingClassifier(estimator=SVC())
bgg.fit(x_train,y_train)
y_pred_bgg=bgg.predict(x_test)
accuracy_score(y_test,y_pred_bgg)
```

#Boosting

```python
#Adaptive Boosting
from sklearn.ensemble import AdaBoostClassifier
ab=AdaBoostClassifier()
ab.fit(x_train,y_train)
y_pred_ab=ab.predict(x_test)
accuracy_score(y_test,y_pred_ab)
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
accuracy_score(y_test,y_pred_gb)
```

```python
#XG Boosting
import xgboost as xgb
xg_model=xgb.XGBClassifier()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
xg_model.fit(x_train,y_train)
y_pred_xg=xg_model.predict(x_test)
accuracy_score(y_test,y_pred_xg)
```

Hyper Parameter Tuning

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_rf)
```

```python
#Bagging Classifier
from  sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(n_estimators=100,random_state=42)
bg.fit(x_train,y_train)
y_pred_bg=bg.predict(x_test)
accuracy_score(y_test,y_pred_bg)
```

```python
#Adaptive Boosting
from sklearn.ensemble import AdaBoostClassifier
ab=AdaBoostClassifier(n_estimators=50,random_state=42)
ab.fit(x_train,y_train)
y_pred_ab=ab.predict(x_test)
accuracy_score(y_test,y_pred_ab)
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=25,random_state=42)
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
accuracy_score(y_test,y_pred_gb)
```

```python
#XG Boosting
import xgboost as xgb
xg=xgb.XGBClassifier()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
xg=xgb.XGBClassifier(n_estimators=100,random_state=42,)
xg.fit(x_train,y_train)
y_pred_xg=xg.predict(x_test)
accuracy_score(y_test,y_pred_xg)
```

```python
#Gradient Boosting
#hyper parameter tuning
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=25,random_state=42,loss="log_loss",criterion="squared_error")
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
accuracy_score(y_test,y_pred_gb)
```
